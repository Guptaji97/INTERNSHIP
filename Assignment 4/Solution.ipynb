You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
import pandas as pd
import selenium 
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
```

```python
driver= webdriver.Chrome(r"C:\Users\amang\chromedriver_win32\chromedriver.exe")
driver.maximize_window()
```

```python
#Getting url
url= "https://www.hostelworld.com/"
driver.get(url)
time.sleep(2)
```

```python
# input in search bar
driver.find_element("xpath","/html/body/div[3]/div/div/div[1]/div[1]/div/div/div[4]/div/div[2]/div/div[1]/div/input").click()

inp=driver.find_element("xpath","/html/body/div[3]/div/div/div[1]/div[1]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input")
inp.send_keys("London")
time.sleep(1)
driver.find_element("xpath",'/html/body/div[3]/div/div/div[1]/div[1]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]').click()
#Search 
driver.find_element("xpath",'/html/body/div[3]/div/div/div[1]/div[1]/div/div/div[4]/div/div[2]/div/div[5]/button').click()
'''inp.send_keys(Keys.DOWN)
inp.send_keys(Keys.ENTER)'''
```

```python
#Names
name=[]
nm= driver.find_elements("xpath",'//h2[@class="title title-6"]')
for i in nm:
    name.append(i.text)
print(len(name))
```

```python
#distance
dist=[]
dstnce= driver.find_elements("xpath",'//span[@class="description"]')
for i in dstnce:
    dist.append(i.text.split("- ")[1])
print(len(dist))
```

```python
#Ratings, Total reviews and Overall reviews
z=[]
ratings=[]
trev=[]
orev=[]
rate= driver.find_elements("xpath",'//div[@class="bottom-rating"]')
for i in rate:
    #print (i.text.split("\n"))
    ratings.append(i.text.split("\n")[0])
    
    if (len(i.text.split("\n"))) == 3: 
        orev.append(i.text.split("\n")[1])
        z.append(i.text.split("\n")[2])
        
    else:
        orev.append("-")
        z.append(i.text.split("\n")[1])
        
for i in z:
    trev.append(i.split(" ")[0])
print(len(ratings),len(orev),len(trev))
```

```python
#Price
d=[]
p=[]
dorm=[]
private=[]
pr= driver.find_elements("xpath",'//div[@class="price-col"]')
c=1
for i in pr:
    price=i.text.split('\n')
    #print(price)
    if c%2==0:
        try:
            d.append(price[1])
        except:
            d.append(price)
        
    else:
        try:
            p.append(price[1])
        except:
            p.append(price)
    c+=1
    
#Private
for i in p:
    try:
        z=(i.split(" "))
        try:
            private.append(z[1])
        except:
            private.append(z)
    except:
        private.append(i)

#dorm        
for i in d:
    try:
        z=(i.split(" "))
        try:
            dorm.append(z[1])
        except:
            dorm.append(z)
    except:
        dorm.append(i)
        
print(len(private),len(dorm))
```

```python
link=[]

for i in range (3,len(dorm)+3):
    try:
        hotel=driver.find_element("xpath",'/html/body/div[3]/div/div/div[1]/div[4]/div/div/div['+str(i)+']/div[3]/div[1]/h2/a')
        #print("done",i)
        
    except NoSuchElementException:
        hotel=driver.find_element("xpath",'/html/body/div[3]/div/div/div[1]/div[4]/div/div/div['+str(i)+']/div[2]/div[1]/h2/a')
        #print("done",i)
        
    link.append(hotel.get_attribute("href"))
print(len(link))

```

```python
description=[]
facilities=[]
for j in range (0,len(link)):
    driver.get(link[j])
    time.sleep(2)    
    
    desc=driver.find_element("xpath",'//div[@class="content collapse-content"]')
    description.append(desc.text)
    
    fac=driver.find_element("xpath",'//div[@class="facilities-container"]')
    facilities.append(fac.text)
    
print(len(description),len(facilities))
```

```python
df=pd.DataFrame({"Hostel Name": name, "Distance from city centre": dist, "Ratings":ratings, "Total Review": trev, "Overall Review": orev, "Privates from price": private, "Dorms from price":dorm, "Facilities": facilities, "Property description": description})
df
```
