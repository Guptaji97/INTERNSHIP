```python
import selenium
from selenium import webdriver 
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
driver= webdriver.Chrome(r"C:\Users\amang\chromedriver_win32\chromedriver.exe")
url="https://www.ambitionbox.com/"
driver.get(url)
```

```python
#Salaries
sal=driver.find_element("xpath","/html/body/div[1]/nav/nav/a[4]")
sal.click()
```

```python
#Input job post
inp= driver.find_element("xpath","/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input")
inp.send_keys("Data Scientist")
```

```python
ds= driver.find_element("xpath","/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p")
ds.click()
```

```python
#Company name
name=[]
exp=[]
sal_record=[]
c=0
cn=driver.find_elements("xpath",'//div[@class="company-info"]')
```

```python
for i in cn:
    if c==10:
        break
    c+=1
    #company name
    co=i.text.split("\n")[0]
    
    #experience
    x=i.text.split("\n")[2]
    ex=x.split("(")[0]
    
    #salary record
    rec=x.split("(")[1][:-1]
    
    #a=i.text
    name.append(co)
    exp.append(ex)
    sal_record.append(rec)
```

```python
#avg ctc
avg=[]
c=0
m=driver.find_elements("xpath",'//p[@class="averageCtc"]')
for i in m:
    if c==10:
        break
    c+=1
    avg.append(i.text)
len(avg)
```

```python
#min and max ctc
low=[]
high=[]
c=0
z=0
l=driver.find_elements("xpath",'//div[@class="value body-medium"]')
for i in l:
    if c==20:
        break
    c+=1
    if z%2==0:
        low.append(i.text)
    if z%2!=0:
        high.append(i.text)
    z+=1
len(low)
```

```python
len(high)
```

```python
df=pd.DataFrame({"Company": name, "Experience": exp, "Salary record": sal_record, "Average CTC": avg, "Min CTC": low, "Max CTC": high})
df
```
