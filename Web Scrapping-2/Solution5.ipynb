```python
summary=[]
rate=[]
details=[]
n=1
c=0
```

```python
while c<100:
    #RATINGS
    r=driver.find_elements("xpath",'//div[@class="_3LWZlK _1BLPMq"]')
    for i in r:
        c=len(rate)
        if c==100:
            break
        rate.append(i.text)
        
    #Summary
    smry=driver.find_elements("xpath",'//p[@class="_2-N8zT"]')
    for i in smry:
        c=len(summary)
        if c==100:
            break
        summary.append(i.text)
        
    #Details
    det=driver.find_elements("xpath",'//div[@class="t-ZTKy"]')
    for i in det:
        c=len(details)
        if c==100:
            break
        details.append(i.text)
    
    if c==100:
        break
    
    n+=1
        
    #Next_button
    try:
        nxt=driver.find_element("xpath",'//a[@class="_1LKTO3"]')
        nxt.click()

        
    except:
        nxt=driver.find_elements("xpath",'//a[@class="ge-49M"]')
        for i in nxt:
            if i.text==str(n):
                i.click()

    

        #nxt.click()
        
        
'''nxt=driver.find_elements("xpath",'//a[@class="_1LKTO3"]')
        #nxt.click()
        for i in nxt:
            if i.text=="NEXT":
                i.click()'''
```
