```python
s_chck=driver.find_element("xpath", "/html/body/div[1]/div[3]/div[2]/section[1]/div[2]/div[3]/div[2]/div[2]/label/i")
s_chck.click()
```

```python
#Job Title
title=[]
tle= driver.find_elements("xpath", '//a[@class="title fw500 ellipsis"]')
len(tle)
```

```python
c=0
for i in tle:
    if c==10:
        break    
    title.append(i.text)
    c+=1
title
```

```python
#Company name
company=[]
comp= driver.find_elements("xpath", '//a[@class="subTitle ellipsis fleft"]')
len(comp)
```

```python
c=0
for i in comp:
    if c==10:
        break
    c+=1
    company.append(i.text)
company
```

```python
# Experience
expr=[]
exp= driver.find_elements("xpath", '//li[@class="fleft grey-text br2 placeHolderLi experience"]')
len(exp)
```

```python
c=0
for i in exp:
    if c==10:
        break
    c+=1
    expr.append(i.text)
expr
```

```python
# Location
locn=[]
l=driver.find_elements("xpath",'//li[@class="fleft grey-text br2 placeHolderLi location"]')
len(l)
```

```python
c=0
for i in l:
    if c==10:
        break
    c+=1
    locn.append(i.text)
locn
```

```python
df=pd.DataFrame({"Job Title": title, "Company": company, "Experience required": expr, "Location": locn})
df
```
