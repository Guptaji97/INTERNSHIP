```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
```

```python
driver= webdriver.Chrome(r"C:\Users\amang\chromedriver_win32\chromedriver.exe")
```

```python
url="http://www.naukri.com/"
driver.get(url)
```

```python
# Input job postition
job= driver.find_element("xpath","/html/body/div[1]/div[2]/div[3]/div/div/div[1]/div/div/div/input")
job.send_keys("Data analyst")
```

```python
# Input Location
locn= driver.find_element("xpath","/html/body/div[1]/div[2]/div[3]/div/div/div[5]/div/div/div/input")
locn.send_keys("Bangalore")
```

```python
# Search button
srch= driver.find_element("xpath", "/html/body/div[1]/div[2]/div[3]/div/div/div[6]")
srch.click()
```

```python
#Job Title
title=[]
tle= driver.find_elements("xpath", '//a[@class="title fw500 ellipsis"]')
len(tle)
```

```python
c=0
for i in tle:
    if c==10:
        break    
    title.append(i.text)
    c+=1
title
```

```python
#Company name
company=[]
comp= driver.find_elements("xpath", '//a[@class="subTitle ellipsis fleft"]')
len(comp)
```

```python
c=0
for i in comp:
    if c==10:
        break
    c+=1
    company.append(i.text)
company
```

```python
# Experience
expr=[]
exp= driver.find_elements("xpath", '//li[@class="fleft grey-text br2 placeHolderLi experience"]')
len(exp)
```

```python
c=0
for i in exp:
    if c==10:
        break
    c+=1
    expr.append(i.text)
expr
```

```python
# Salary
sal=[]
s= driver.find_elements("xpath",'//li[@class="fleft grey-text br2 placeHolderLi salary"]')
len(s)
```

```python
c=0
for i in s:
    if c==10:
        break
    c+=1
    sal.append(i.text)
sal
```

```python
# TimeStamp
time=[]
t=driver.find_elements("xpath",'//div[@class="jobTupleFooter mt-20"]')
len(t)
```

```python
c=0
for i in t:
    if c==10:
        break
    c+=1
    a=i.text.split("\n")
    time.append(a[0])
time
```

```python
df=pd.DataFrame({"Job Title": title, "Company": company, "Experience required": expr, "Salary": sal, "TimeStamp": time})
df
```
