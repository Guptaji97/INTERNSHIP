#### Scrapping page 1

```python
#Getting brand details page 1
brand=[]
b=driver.find_elements("xpath",'//div[@class="_2WkVRV"]')
```

```python
for i in b:
    a=len(brand)
    if a==100:
        break
    brand.append(i.text)
len(brand)
```

```python
#Getting description page 1
des=[]
d=driver.find_elements("xpath",'//a[@class="IRpwTa"]')
```

```python
for i in d:
    a=len(des)
    if a==100:
        break
    des.append(i.text)
len(des)
```

```python
#Getting price details page 1
price=[]
p=driver.find_elements("xpath",'//div[@class="_30jeq3"]')
```
