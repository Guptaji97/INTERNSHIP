```python
#job title
title=[]
c=0
t=driver.find_elements("xpath",'//a[@class="title noclick"]')
```

```python
for i in t:
    if c==10:
        break
    c+=1
    title.append(i.text)
len(title)
```

```python
#Job descripption
desc=[]
c=0
d=driver.find_elements("xpath",'//p[@class="company body-medium"]')
```

```python
for i in d:
    if c==10:
        break
    c+=1
    desc.append(i.text)
len(desc)
```

```python
#Ratings
rating=[]
c=0
r=driver.find_elements("xpath",'//span[@class="body-small"]')
```

```python
for i in r:
    if c==10:
        break
    c+=1
    rating.append(i.text)
len(rating)
```

```python
#Experience and Qualification
exp=[]
qual=[]
c=0
e=driver.find_elements("xpath",'//div[@class="job-basic-info show-flex"]')
```

```python
for i in e:
    if c==10:
        break
    c+=1
    a=i.text.split("\n")[0]
    b=i.text.split("\n")[1]
    exp.append(a)
    qual.append(b)
    qual.append
len(exp)
```
