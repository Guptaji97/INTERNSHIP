```python
import selenium
from selenium import webdriver 
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver= webdriver.Chrome(r"C:\Users\amang\chromedriver_win32\chromedriver.exe")
```

```python
url= "https://www.amazon.in/"
driver.get(url)
```

```python
# input in search bar
inp=driver.find_element("xpath","/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
inp.send_keys("Laptop")
```

```python
inp=driver.find_element("xpath","/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
inp.click()
```

```python
#select intel I7
chck=driver.find_elements("xpath",'//span[@class="a-list-item"]')
for i in chck:
    a= i.text
    if a == "Intel Core i7":
        print(i.text + " Selected")
        i.click()
        break
```

```python
# getting title
title=[]
c=0
t=driver.find_elements("xpath",'//span[@class="a-size-medium a-color-base a-text-normal"]')
```

```python
for i in t:
    if c<10:
        title.append(i.text)
    c+=1
len(title)
```

```python
#price
price=[]
c=0
p=driver.find_elements("xpath",'//span[@class="a-price-whole"]')
```

```python
for i in p:
    if c<10:
        price.append(i.text)
    c+=1
len(price)
```

```python
#Rating
titles_rat=driver.find_elements('xpath',"//div[@class='a-row a-size-small']/span")
s=[]
for i in titles_rat:
    s.append(i.get_attribute("aria-label"))
```

```python
# extratcing ratings from the unorganised rating tag
c=0
rating=[]

for i in range(0,len(s)):
    
    if c<20:
        if i%2==0:
            rating.append(s[i][0:3])
    c+=1
```

```python
rating
```

```python
df=pd.DataFrame({"Title":title, "Price":price, "Rating": rating})
df
```
