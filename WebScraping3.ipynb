```python
# get link of all the products available on the current page
link= []

l=driver.find_elements("xpath",'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
for i in l:
    link.append(i.get_attribute("href"))
print(len(link))
```

```python
for i in page:
    driver.get(i)
    time.sleep(3)

#get url of all products
    l=driver.find_elements("xpath",'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in l:
        link.append(i.get_attribute("href"))
print(len(link), "total products.")
```

```python
brand=[]
prod=[]
price=[]
returnable=[]
delvry=[]
avail=[]
url=[]
c=0


for i in link:
    driver.get(i)
    time.sleep(3)
    
# Get Brand name
    b=driver.find_elements("xpath",'//a[@class="a-link-normal"]')
    z=[]
    emp=[]
    for i in range(0,len(b)):
        z.append(b[i].text.split())
    
        if (emp in z):     # remove empty elements from list
            z.remove(emp)
#print(z)
    print("getting details of product no",c,"out of", len(link))
    
    for i in range(0,len(z)): 
        if ("Visit" in z[i][0]):
            ind=i
            brand.append(" ". join(z[ind][2:-1]))
            break
            
        elif ("Brand:" in z[i][0]):
            ind=i
            brand.append(" ". join(z[ind][1:]))
            break         
    
    prod.append(driver.find_element("xpath",'//span[@class="a-size-large product-title-word-break"]').text)    #product details
    
    price.append(driver.find_element("xpath",'//span[@class="a-price-whole"]').text) #price
    
#Return/exchange
    rtrn= driver.find_elements('xpath','//a[@class="a-size-small a-link-normal a-text-normal"]')
    r=[]
    for i in rtrn:
        r.append(i.text.split())
    
    for k in range(0,len(r)):
        #print("inside loop",k)
        if ("Replacement" in r[k][-1]) or ("Returnable" in r[k][-1]):
            #print("found")
            #ind=k
            returnable.append(" ".join(r[k]))
            break
    if len(returnable)==c:
        returnable.append("-")    #if nothing is appended in above line then append it with "-".
    #print(len(returnable))
    
#availability
    try:
        avail.append(driver.find_element("xpath",'//span[@class="a-size-medium a-color-success"]').text)
        
    except:
        avail.append("-")    
    
    
#Delivery
    delvry.append(driver.find_element("xpath",'//span[@class="a-text-bold"]\n').text)
   
    c+=1

print("\n\nScrapping product details is successful.")
```

```python
# replacing empty elements with "-"

list1= [brand, prod, price, returnable, delvry, avail, link]

for i in range(len(list1)):
    for j in range(len(list1[i])):
        if (list1[i][j]==""):
            print("at index [{0} {1}]".format(i,j), end=" ")
            list1[i][j]="-"
            print(list1[i][j], 'is appended')
```

```python
print(len(brand), len(prod), len(price), len(returnable), len(delvry), len(avail), len(link))
```

```python
df= pd.DataFrame({"Brand": brand, "Product details": prod, "Price":price, "Returnable":returnable, "Availability": avail, 
                  "Delivery": delvry, "URL": link})
df
```

#### 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page.
```python
brand=[]
model=[]
color=[]

for i in driver.find_elements("xpath",'//div[@class="_4rR01T"]'):
    brand.append(i.text.split()[0])     #brand
    
    a=" ".join(i.text.split()[1:])
    model.append(a.split("(")[0])    #model
    
    #color
    try:
        color.append(a.split("(")[1].split(",")[0])
        
    except IndexError:
        color.append("-")
```

```python
ram=[]
rom=[]
sort=[]
display=[]
pcam=[]
scam=[]
bat=[]

a= driver.find_elements("xpath",'//ul[@class="_1xgFaf"]')
for i in a:
    sort.append(i.text.split("\n"))
    
for i in sort:
    ram.append(i[0].split('|')[0])
    rom.append(", ".join(i[0].split(' | ')[1:]))
    
    display.append((" ").join(i[1].split()[:4]))
    
    pcam.append(i[2].split("|")[0])
    try:
        scam.append(i[2].split("| ")[1].split()[0])
    except IndexError:
        scam.append("-")
        
    bat.append(" ".join(i[3].split()[:2]))
```

#### 6. Write a program to scrap details of all the funding deals for second quarter (i.e Jan 21 – March 21) from trak.in

```python
import pandas as pd
import selenium
import time
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
```

```python
driver= webdriver.Chrome(r'C:\Users\amang\chromedriver_win32\chromedriver.exe')
driver.maximize_window()

url="https://trak.in/india-startup-funding-investment-2015/"
driver.get(url)
time.sleep(3)
```

```python
april=[]
quat=[]
for i in driver.find_elements("xpath",'//td[@class="column-2"]'):
    #print(i.text.split("/"))
    z=(i.text.split("/"))
    
    if (z[1]=="01") or (z[1]=="02") or (z[1]=="03") and (z[2]=="2021"):
        quat.append(i.text)
        
    elif (z[1]=="04") and (z[2]=="2021"):
        april.append(i.text)
               
    else:
        break
        
    #elif((i.text.split("/")[1]))
```

```python
sr=[]
name=[]
indus=[]
sub=[]
city=[]
investor=[]
typ=[]
usd=[]
for j in range(1,10):
    if j==2:
        continue
    sv=driver.find_elements("xpath",'//td[@class="column-'+str(j)+'"]')
    
    for i in range(len(sv)):
        if i<len(april):
            continue
        
        elif i<(len(april)+len(quat)):
            
            if j==1:
                sr.append(sv[i].text)
            
            if j==3:
                name.append(sv[i].text)
            
            if j==4:
                indus.append(sv[i].text)
            
            if j==5:
                sub.append(sv[i].text)
                
            if j==6:
                city.append(sv[i].text)
                
            if j==7:
                investor.append(sv[i].text)
                
            if j==8:
                typ.append(sv[i].text)
                
            if j==9:
                usd.append(sv[i].text)
                      
        else:
            break
```

```python
len(sr), len(quat), len(name), len(indus), len(sub), len(city), len(investor), len(typ), len(usd)
```

```python
df= pd.DataFrame({"Sr. No.": sr, "Date":quat, "Startup Name": name, "Industry": indus, "Sub-vertical": sub, "Location":city,
                 "Investors' Name": investor, "Investment type": typ, "Amount in USD":usd})
df.set_index("Sr. No.",inplace=True)
df
```

```python
driver.close()
```

#### 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
import pandas as pd
import selenium
import time
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
```

```python
driver= webdriver.Chrome(r'C:\Users\amang\chromedriver_win32\chromedriver.exe')
driver.maximize_window()
```

```python
url="https://www.youtube.com/watch?v=qySOXAJW12g"
driver.get(url)

time.sleep(20)

driver.execute_script("window.scrollTo(0,500);") # scroll little down to load comment section
time.sleep(5)
```

```python
def scroll():
    #Scroll through all the comments to load them
    c=0
    i=0
    while i<=20:
        i+=1
        c+=10000
        driver.execute_script("window.scrollTo(0,"+str(c)+");")
        time.sleep(5)
```

```python
scroll()    #calling function
while True:
    name=[]
    comments=[]
    upvote=[]
    ts=[]
    
    #get names
    for i in driver.find_elements("xpath",'//span[@class=" style-scope ytd-comment-renderer"]'):
        name.append(i.text)
        
    #get time stamp
    for i in driver.find_elements("xpath",'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]'):    
        ts.append(i.text.split("(")[0])
        
    #get comments
    for i in driver.find_elements("xpath",'//yt-formatted-string[@class="style-scope ytd-comment-renderer"]'):
        comments.append(i.text)
        
    # get upvotes    
    uv=driver.find_elements("xpath",'//span[@class="style-scope ytd-comment-action-buttons-renderer"]')
    for i in range (len(uv)):
        if i%2==0:
            continue
   
        upvote.append(uv[i].text)
        
    if len(name)>=500:
        break     # break after scrapping minimum 500 comments
        
    else:
        print(len(name), "no of data is scrapped, repeating the process...")
        scroll()    #scroll and try again till atleast 500 comments is scrapped
```

```python
print(len(name), len(comments), len(upvote), len(ts))
```

```python
df= pd.DataFrame({"Name": name, "Comment":comments, "Number of upvotes":upvote, "Timestamp":ts})
df
```

```python
link=[]
name=[]
distance=[]
rate=[]
total=[]
over=[]
p=[]
d=[]
private=[]
dorm=[]

while True:
#scroll to load all hostels
    driver.execute_script("window.scrollBy(0,10000);")
    
    for i in driver.find_elements("xpath",'//h2[@class="title title-6"]/a'):
        link.append(i.get_attribute("href"))    #get link of all the hostels at every page
        
        name.append(i.text)    #name of hostel
    time.sleep(2)
    
    for i in driver.find_elements("xpath",'//span[@class="description"]'):
        distance.append(i.text.split("- ")[1].split("from")[0])    #distance from city center
        
    r=driver.find_elements("xpath",'//div[@class="rating rating-summary-container big"]')
    for i in range(len(r)):
        rate.append(r[i].text.split("\n")[0])    # ratings
        total.append(r[i].text.split("\n")[-1].split(" ")[0])    #total review
    
        if ((len(r[i].text.split("\n")))<3) or (r[i].text.split("\n")[1] == "Rating"):    #overall ratings
            over.append("-")    
    
        else:
            over.append(r[i].text.split("\n")[1])    
            
    pr=driver.find_elements("xpath",'//div[@class="price-col"]')
    for i in range(len(pr)):
        price.append(pr[i].text.split("\n"))    #contains both private and dorm prices alternatively
    
        if i%2==0:    #sort out dorm 
            try:
                dorm.append(price[i][1])
            
            except:
                dorm.append(price[i][0])
            
        else:    #sort out private
            try:
                private.append(price[i][1])
            
            except:
                private.append(price[i][0])
            
    for i in range(len(private)):
        if (len(private[i].split(" "))>1) and (len(private[i].split(" "))<3):
            private[i]=(private[i].split(" ")[1])    #choose correct price where more than 2 prices are mentioned
        
    for i in range(len(dorm)):
        if (len(dorm[i].split(" "))>1) and (len(dorm[i].split(" "))<3):
            dorm[i]=(dorm[i].split(" ")[1])    #choose correct price where more than 2 prices are mentioned
    
    #next page
    try:
        driver.find_element("xpath",'/html/body/div[3]/div/div/div[2]/div[4]/div/div/div[33]/div[4]').click()
        time.sleep(3)
    
    except e:
        print("No such Element found, It seems u have scrapped all the available data.")
        break
        
print(len(link), "hostels' link has been scrapped.")
```

```python
facility=[]
desc=[]
c=0
for i in link:
    print("Scrapping hostel {0}".format(c))
    driver.get(i)
    time.sleep(3)
    driver.execute_script("window.scrollBy(0,2000);")
    
    a=[]
    for j in driver.find_elements("xpath",'//ul[@class="facilities"]/li'):
        a.append(j.text)
    facility.append(a)    # facilities
    
    try:
        d=driver.find_element("xpath",'//div[@class="content collapse-content"]')
        
    except e:
        d=driver.find_element("xpath",'//div[@class="content"]')
    desc.append(d.text)    #description
    
    c+=1
```
