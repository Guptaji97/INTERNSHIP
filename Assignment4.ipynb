#### 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos. You need to find following details: A) Rank B) Name C) Artist D) Upload date E) Views

```python
import pandas as pd
import selenium
import time
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
```

```python
driver= webdriver.Chrome(r'C:\Users\amang\chromedriver_win32\chromedriver.exe')
driver.maximize_window()

url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"    #url
driver.get(url)
time.sleep(5)
```

```python
rank=[]
name=[]
art=[]
date=[]
view=[]

for i in range(1,31):    #get rows
    for j in range(1,6):     #get columns
        a=driver.find_element("xpath",'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr['+str(i)+']/td['+str(j)+']')
        
        if j%5==1:
            rank.append(a.text)    #rank
            
        if j%5==2:
            name.append(a.text.split("[")[0])    #name
            
        if j%5==3:
            art.append(a.text)    #artist
            
        if j%5==4:
            view.append(a.text)    #date
            
        if j%5==0:
            date.append(a.text)    #views

print(len(rank), len(name), len(art), len(date), len(view))        
```

```python
df= pd.DataFrame({"Rank":rank, "Name": name, "Artist": art, "Publication date": date, "Views": view})
df=df.set_index("Rank")
df
```

```python
driver.close()
```

#### 5. Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details: A) Repository title B) Repository description C) Contributors count D) Language used.

```python
import pandas as pd
import selenium
import time
import warnings
from selenium import webdriver
warnings.filterwarnings("ignore")
```

```python
driver= webdriver.Chrome(r'C:\USers\amang\chromedriver_win32\chromedriver.exe')
driver.maximize_window()

url="https://github.com/"
driver.get(url)
time.sleep(5)
```

```python
a=driver.find_elements("xpath",'//li[@class="HeaderMenu-item position-relative flex-wrap flex-justify-between flex-items-center d-block d-lg-flex flex-lg-nowrap flex-lg-items-center js-details-container js-header-menu-item"]')
for i in a:
    if (i.text)=='Open Source':
        i.click()
        break
        
driver.find_element("xpath",'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a').click()
time.sleep(3)
```

```python
link=[]
for i in driver.find_elements("xpath",'//h1[@class="h3 lh-condensed"]/a'):
    link.append(i.get_attribute("href"))
```

```python
title=[]
des=[]
count=[]
lang=[]
for i in link:
    driver.get(i)
    time.sleep(3)
    
    # Repository title
    a= (driver.find_element('xpath','//div[@class=" d-flex flex-wrap flex-items-center wb-break-word f3 text-normal"]').text)
    title.append(a.split("/\n")[1].split("\n")[0])
    
    des.append(driver.find_element("xpath",'//p[@class="f4 my-3"]').text)    #description

    cnt= driver.find_elements("xpath",'//span[@class="Counter"]')    #counter
    count.append(cnt[-1].text)
        
    l=[]
    for i in driver.find_elements("xpath",'//span[@class="color-fg-default text-bold mr-1"]'):
        l.append(i.text)
    lang.append(l)
```

```python
len(title), len(des), len(count), len(lang)
```

```python
df= pd.DataFrame({"Repository title": title, "Repository description": des, "Contributors count": count, "Languages used": lang})

for i in df.columns:                 # fill empty values with "-"
    for c in range (len(title)):
        if len(df[i][c])==0:
            df[i][c]="-"

df
```

#### 9. Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455/ . You have to find the following details: A) Name B) Year span C) Genre D) Run time E) Ratings F) Votes

```python
import pandas as pd
import selenium
import time
import warnings
warnings.filterwarnings("ignore")
from selenium import webdriver
```

```python
driver=webdriver.Chrome(r"C:\Users\amang\chromedriver_win32\chromedriver.exe")
driver.maximize_window()

url="https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(5)
```

```python
#scroll down
driver.execute_script("window.scrollBy(0,20000);")
```

```python
name=[]
year=[]
genre=[]
runtime=[]
rate=[]
vote=[]
c=0

for i in driver.find_elements("xpath",'//h3[@class="lister-item-header"]'):    #get name and year of genre
    if len(i.text.split(" ")[-1])==11:
        name.append(" ".join(i.text.split(" ")[1:-1]))
        year.append(i.text.split(" ")[-1])
        
    else:
        name.append(" ".join(i.text.split(" ")[1:-2]))
        year.append(" ".join(i.text.split(" ")[-2:]))

for i in driver.find_elements("xpath",'//p[@class="text-muted text-small"]'):
    if c%3==0:
        runtime.append(i.text.split("|")[1])    #runtime
        genre.append(i.text.split("|")[2])     #genre
    
    elif c%3==2:
        vote.append(i.text.split(": ")[1])     #votes
    c+=1

for i in driver.find_elements("xpath",'//div[@class="ipl-rating-star small"]'):    #ratings
    rate.append(i.text)
```

```python
print(len(name),len(year),len(genre),len(runtime),len(rate), len(vote))
```

#### 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ . You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year

```python
import pandas as pd
import selenium
import time
import warnings
from selenium import webdriver
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome(r"C:\Users\amang\chromedriver_win32\chromedriver.exe")
driver.maximize_window()

url="https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(5)

driver.find_element("xpath",'/html/body/table[1]/tbody/tr/td[2]/span[2]').click()    #click "view all datasets"
time.sleep(3)
```

```python
name=[]
data=[]
task=[]
types=[]
instance=[]
attribute=[]
year=[]

for i in range (2,624):
#name
    nm=driver.find_element("xpath",'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[1]/table/tbody/tr/td[2]/p/b/a')
    name.append(nm.text)
#datatype
    dt=driver.find_element("xpath",'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[2]/p')
    data.append(dt.text)
#task
    tsk=driver.find_element("xpath",'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[3]/p')
    task.append(tsk.text)
#attribute type
    ty=driver.find_element("xpath",'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[4]/p')
    types.append(ty.text)
#instance
    inst=driver.find_element("xpath",'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[5]/p')
    instance.append(inst.text)
#attributes nos
    attr=driver.find_element("xpath",'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[6]/p')
    attribute.append(attr.text)
#year
    yr=driver.find_element("xpath",'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[7]/p')
    year.append(yr.text)
```

```python
len(name), len(data), len(task), len(types), len(instance), len(attribute), len(year)
```

```python
df=pd.DataFrame({"Dataset name":name, "Data type":data, "Task":task, "Attribute types":types, "no. of Instances":instance, "no. of Attributes":attribute, "Year":year})

for i in df.columns:                 # fill empty values with "-"
    for c in range (len(name)):
        if (df[i][c])==" ":
            df[i][c]="-"
            
df
```

```python
driver.close()
```
