```python
''' Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) 
from https://presidentofindia.nic.in/former-presidents.htm
'''
from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://presidentofindia.nic.in/former-presidents.htm"
req=requests.get(url)
soup=BeautifulSoup(req.content)
```

```python
#Concept
s=soup.find("div",class_="presidentListing").text
n=[]
print(s)
a=s.index("http")
s=s[0:a]
print("After Slicing",s)
s.index("1935")-len(s)
n.append(s[0:-61])
print("Only name",n)
```

```python
name=[]
term=[]
for i in soup.find_all("div",class_="presidentListing"):
    a=i.text
    #For name
    if "birth" in a:
        b=a.index("birth")
        name.append(a[1:b-2])
        
    else:
        b=a.index("Term")-13
        name.append(a[1:b])
    
    #for term
    if "http" in a:
        t=a.index(":")
        h=a.index("http")
        term.append(a[t+2:h-1])
        
    else:
        t=a.index(":")
        term.append(a[t+2:-1])
    
    '''print(a.count(a))
    if c in a:
        print('yes')
        print(a)
        name.append(a[0:-61])
    else:
        name.append(a)'''
name
```
