```python
'''Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
release) and make data frame'''

from bs4 import BeautifulSoup
import requests
import pandas as pd

url="https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=461131e5-5af0-4e50-bee2-223fad1e00ca&pf_rd_r=NHV9HWT5P7ZEM6DGVBKC&pf_rd_s=center-1&pf_rd_t=60601&pf_rd_i=india.toprated&ref_=fea_india_ss_toprated_india_tr_india250_sm"
```

```python
req=requests.get(url)
soup=BeautifulSoup(req.content)
```

```python
name=[]
year=[]
c=0
for i in soup.find_all("td",class_="titleColumn"):
    a=i.text
    if c<10:
        name.append(a[16:-8])
    elif 9<c<100:
        name.append(a[17:-8])
    else:
        name.append(a[18:-8])
    year.append(a[-6:-2])
    c+=1
name
```

```python
year
```

```python
ratings=[]
for i in soup.find_all("td",class_="ratingColumn imdbRating"):
    a=i.text.split()[0]
    ratings.append(a)
ratings
```

```python
df=pd.DataFrame({"Movie name":name, "Year":year, "Ratings":ratings})
df
```
