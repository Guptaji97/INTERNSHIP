```python
'''Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location 
iv) Ratings
v) Image URL'''

from bs4 import BeautifulSoup
import requests
import pandas as pd

url="https://www.dineout.co.in/delhi-restaurants/buffet-special"
req=requests.get(url)
soup=BeautifulSoup(req.content)
```

```python
#name
s=soup.find("a",class_="restnt-name ellipsis")
s.text
```

```python
#Cusine
s=soup.find("span",class_="double-line-ellipsis").a
s.text
```

```python
#location
s=soup.find("div",class_="restnt-loc ellipsis")
s.text
```

```python
#Ratings
s=soup.find("div",class_="restnt-rating rating-4")
s.text
```

```python
#ImageURL
s=soup.find("img",class_="no-img")
s["data-src"]
```

```python
#name
name=[]
for i in soup.find_all("a",class_="restnt-name ellipsis"):
    name.append(i.text)
name
```

```python
#Cusine
cus=[]
for i in (soup.find_all("span",class_="double-line-ellipsis")):
    cus.append(i.text.split("|")[1])
cus
```

```python
#location
loc=[]
for i in soup.find_all("div",class_="restnt-loc ellipsis"):
    loc.append(i.text)
loc
```

```python
#Ratings
rt=[]
for i in soup.find_all("div",class_="restnt-rating rating-4"):
    rt.append(i.text)
rt
```

```python
#ImageURL
img=[]
for i in soup.find_all("img",class_="no-img"):
    img.append(i["data-src"])
img
```

```python
df=pd.DataFrame({"Restaurant:":name, "Cuisine":cus, "Location":loc, "Ratings":rt, "Image URL":img})
df
```
