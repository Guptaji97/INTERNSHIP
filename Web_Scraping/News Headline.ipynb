```python
''' Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
i) Headline
ii) Time
iii) News Link'''

from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://www.cnbc.com/world/?region=world"
req=requests.get(url)
soup=BeautifulSoup(req.content)
```

```python
head=[]
for i in soup.find_all("a",class_="LatestNews-headline"):
    head.append(i.text)
head
```

```python
time=[]
for i in soup.find_all("time",class_="LatestNews-timestamp"):
    time.append(i.text)
time
```

```python
link=[]
for i in soup.find_all("a",class_="LatestNews-headline"):
    link.append(i['href'])
link
```

```python
df=pd.DataFrame({"Latest News Headline":head, "TimeStamp":time, "Link":link})
df
```
