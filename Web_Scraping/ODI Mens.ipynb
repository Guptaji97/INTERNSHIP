```python
'''5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.


NOTE: I got 20 nos of best data and using that though i can display only 10 using for loop'''

from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
req=requests.get(url)
soup=BeautifulSoup(req.content)
```

```python
#For new zealand
n=soup.find("span",class_="u-hide-phablet")
name=[n.text]

m=soup.find("td",class_="rankings-block__banner--matches")
matches=[m.text]

pt=soup.find("td",class_="rankings-block__banner--points")
points=[pt.text]

rt=soup.find("td",class_="rankings-block__banner--rating u-text-right")
ratings=[rt.text.split()[0]]

#for other Team names
n=soup.find_all("td",class_="table-body__cell rankings-table__team")
#team.append(name.text.replace("\n","")[:-3])
for i in range (0,19):
    name.append(n[i].text[0:-4].replace("\n",""))
#break
name

#for other Team matches and points
no=soup.find_all("td",class_="table-body__cell u-center-text")
for i in range (0,38):
    if i%2==0:
        matches.append(no[i].text)
    else:
        points.append(no[i].text)

#for other ratings
rt=soup.find_all("td",class_="table-body__cell u-text-right rating")
for i in range (0,19):
    ratings.append(rt[i].text)
```

```python
df=pd.DataFrame({"Team name":name, "No. of Matches":matches, "Points": points, "Ratings": ratings})
df
```
