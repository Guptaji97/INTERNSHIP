```python
''' Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
b) Top 10 women’s ODI Batting players along with the records of their team and rating.
c) Top 10 women’s ODI all-rounder along with the records of their team and rating.
'''
from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://www.icc-cricket.com/rankings/womens/team-rankings/odi"
req=requests.get(url)
soup=BeautifulSoup(req.content)
```

```python
#all name
name=[]
n=soup.find_all("span",class_="u-hide-phablet")
for i in range(0,11):
    name.append(n[i].text)
#name
```

```python
#Details of rank 1

mo=soup.find("td",class_="rankings-block__banner--matches")
matches=[mo.text]

po=soup.find("td",class_="rankings-block__banner--points")
points=[po.text]

ro=soup.find("td",class_="rankings-block__banner--rating u-text-right")
ratings=[ro.text.split()[0]]

# matches and points from index 1
m=soup.find_all("td",class_="table-body__cell u-center-text")
for i in range(0,20):
    if i%2==0:
        matches.append(m[i].text)
    else:
        points.append(m[i].text)
#print(matches)
#print(points)

#Ratings from indexx 1
r=soup.find_all("td",class_="table-body__cell u-text-right rating")
for i in range(0,10):
    ratings.append(r[i].text)
#ratings
```

```python
df=pd.DataFrame({"Team name":name, "No. of Matches":matches, "Points": points, "Ratings": ratings})
df
```
