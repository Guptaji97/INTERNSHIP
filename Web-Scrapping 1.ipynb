#### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#getting url
url="https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&count=100"
req=requests.get(url)
soup=BeautifulSoup(req.content)
```

```python
#year and name
name=[]
year=[]
for i in soup.find_all('h3', class_="lister-item-header"):
    if(len(i.text.split("\n")[3])>6):   #split method to split name and year
# movie sequence e.g. Joker(1). here, (1) is mentioned in year section so i used concat if sequence of movie is mentioned too
        name.append((i.text.split("\n")[2]) + (i.text.split("\n")[3][:3]))
        year.append(i.text.split("\n")[3][5:-1])
    else:
        name.append(i.text.split("\n")[2])
        year.append(i.text.split("\n")[3][1:-1])
```

```python
#get ratings
rat=[]
for i in soup.find_all('div', class_="inline-block ratings-imdb-rating"):
    rat.append(i.text.split("\n")[2])
```

```python
print(len(name), len(year), len(rat))
```

```python
df= pd.DataFrame({"Movie name": name, "Rating": rat, "Year": year})
df
```

#### 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=461131e5-5af0-4e50-bee2-223fad1e00ca&pf_rd_r=ZGFMHWR05N1CA84RY586&pf_rd_s=center-1&pf_rd_t=60601&pf_rd_i=india.toprated&ref_=fea_india_ss_toprated_india_tr_india250_sm"
req= requests.get(url)
soup= BeautifulSoup(req.content)
```

```python
#name and year
name=[]
year=[]
c=0                                                 #counter
for i in soup.find_all("td", class_="titleColumn"):
    if c>=100:
        break                                       #break the loop after getting 100 list elements
    name.append(i.text.split("\n")[2])
    year.append(i.text.split("\n")[-2][1:-1])
    c+=1
```

```python
#ratings
rat=[]
c=0
for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    if c>=100:
        break
    rat.append(i.text.split("\n")[1])
    c+=1
```

```python
print(len(name), len(year), len(rat))
```

```python
df= pd.DataFrame({"Name": name, "Rating": rat, "Year": year})
df
```

#### 4) Write a python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://presidentofindia.nic.in/former-presidents.htm"
req= requests.get(url)
soup= BeautifulSoup(req.content)
```

```python
name=[]
term=[]
for i in soup.find_all("div",class_="presidentListing"):
    #a=i.text.split("\n")[1]                              #first split is to split term and name with birth
    #print(a.split("(")[0])                               #Second split is to split name and birth
    name.append(i.text.split("\n")[1].split("(")[0])      # both split when combined can be written in this way
    
# term
    term.append(i.text.split("\n")[2].split(":")[1])      # with same logic we can get the term
```

#### 5) Write a python program to scrape cricket rankings from http://icc-cricket.com.
```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

######                              a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
url="https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
req= requests.get(url)
soup= BeautifulSoup(req.content)
```

```python
name=[]
n= soup.find_all("span",class_="u-hide-phablet")
for i in range (0,10):
    name.append(n[i].text)
```

```python
match=[]
points=[]

m1=soup.find("td",class_="rankings-block__banner--matches")
match.append(m1.text)                                              # get first team matches

p=soup.find("td", class_="rankings-block__banner--points")
points.append(p.text)                                              # get first team points

m= soup.find_all("td",class_="table-body__cell u-center-text")
for i in range(0,18):
    if (i%2==0):                                                   # appends data in match and points column one by one.
        match.append(m[i].text)                                    
    
    else:
        points.append(m[i].text)
```

```python
rate=[]
r1= soup.find('td',class_="rankings-block__banner--rating u-text-right")
rate.append(r1.text.split("\n")[1].split(" ")[-1])                         # get rating of first team

r= soup.find_all('td',class_="table-body__cell u-text-right rating")
for i in range(0,9):
    rate.append(r[i].text)

```

```python
len(name),len(match),len(points),len(rate)
```

```python
df= pd.DataFrame({"Team name":name, "Matches":match, "Points":points, "Ratings":rate})
df
```

###### a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.icc-cricket.com/rankings/womens/team-rankings/odi"
req= requests.get(url)
soup= BeautifulSoup(req.content)
```

```python
name=[]
n= soup.find_all("span",class_="u-hide-phablet")   #getting team names
for i in range(0,10):
    name.append(n[i].text)
```

```python
match=[]
points=[]
m= soup.find("td", class_="rankings-block__banner--matches")   #get first team matches 
match.append(m.text)

p= soup.find("td", class_="rankings-block__banner--points")    # get total points
points.append(p.text)

m= soup.find_all("td",class_="table-body__cell u-center-text") # get remaining team matches
for i in range(0,18):
    if i%2==0:
        match.append(m[i].text)
    else:
        points.append(m[i].text)
```

```python
rate=[]
r= soup.find("td", class_="rankings-block__banner--rating u-text-right") #get rating of first team
rate.append(r.text.split("\n")[1])

r= soup.find_all("td",class_="table-body__cell u-text-right rating")     #get rating of other teams
for i in range(0,9):
    rate.append(r[i].text)
```

```python
print(len(name),len(match),len(points),len(rate))
```

```python
df= pd.DataFrame({"Name": name, "Matches": match, "Points": points, "Rating":rate})
df
```

#### 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world : i) Headline ii) Time iii) News Link

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.cnbc.com/world/?region=world"
req=requests.get(url)
soup= BeautifulSoup(req.content)
```

```python
head=[]
link=[]
h=soup.find_all('a',class_="LatestNews-headline")   
for i in h:
    head.append(i.text)                #get news headlines
    link.append(i['href'])             #get news link
```

```python
time=[]
for i in soup.find_all("span",class_="LatestNews-wrapper"):
    time.append(i.text)                #get timestamp
```

```python
print(len(head),len(time), len(link))
```

```python
df= pd.DataFrame({"Headlines":head, "Timestamp":time, "Link":link})
df
```

```python
print(len(name), len(cus),len(loc),len(rate),len(img))
```

#### 10) Write a python program to scrape the details of top publications from Google Scholar from  https://scholar.google.com/citations?view_op=top_venues&hl=en i) Rank ii) Publication iii) h5-index iv) h5-median

```python
url="https://scholar.google.com/citations?view_op=top_venues&hl=en"
req= requests.get(url)
soup= BeautifulSoup(req.content)
```

```python
rank=[]
for i in soup.find_all("td",class_="gsc_mvt_p"):
    rank.append(i.text)
```

```python
publication=[]
for i in soup.find_all("td",class_="gsc_mvt_t"):
    publication.append(i.text)
```

```python
index=[]
median=[]
c=0
for i in soup.find_all("td",class_="gsc_mvt_n"):
    if c%2==0:
        index.append(i.text)
    else:
        median.append(i.text)
    c+=1
```

```python
print(len(rank),len(publication),len(index), len(median))
```

```python
df=pd.DataFrame({"Rank":rank, "Publication":publication, "h5-index":index, "h5-median":median})
df= df.set_index("Rank")
df
```
